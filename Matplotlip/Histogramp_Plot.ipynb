## Matplotlib Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(10,60,50)
x
```

```python
a = [22, 23, 48, 36, 56, 16, 17, 59, 30, 46, 59, 31, 40, 12, 34, 59, 32,
       37, 59, 13, 17, 22, 17, 24, 59, 37, 59, 39, 41, 52, 32, 32, 22, 26,
       13, 33, 21, 11, 26, 41, 57, 31, 21, 11, 16, 30, 32, 47, 30, 10]
print(a)
```

```python
plt.hist(a, label="Points", cumulative=-1, histtype="barstacked", log=True)
plt.legend()
plt.title("WSCUBE")
plt.xlabel("Number")
plt.ylabel("Count")
# plt.xvline(30, color='g')

# Show grid line
plt.grid()
plt.show()
```

- Use Cumulative to revers
- To Change Type use histtype
- TO Convert Log use log= True
