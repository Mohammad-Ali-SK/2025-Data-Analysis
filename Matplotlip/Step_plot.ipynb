## Matplotlib Step Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
plt.step(x,y,color="r",marker="*", label = "Python")


plt.legend()
plt.title("Python")
plt.xlabel("x -axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

- Use ms = 10 (To change marker size)
- Use mfc = "g" (To chage marker color to green)
