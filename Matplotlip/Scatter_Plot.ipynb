## Matplotlib Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(1,11)
y = [i*i for i in a]
x = a.tolist()
print(x)
print()
print(y)
```

```python
plt.scatter(x,y, label = "point", cmap="BrBG")
plt.colorbar()
plt.legend(fontsize = 8)
plt.title("Scatter Plot", fontsize = 10)
plt.xlabel("Day's")
plt.ylabel("Count")
plt.show()
```
