## Pie Chart in Python

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(6,5))

x = [24,76,88,99]
y = ["c","c++", "Java","Python"]

ex  = [0.2,0.0,0.0,0.00]


plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=True)
plt.legend(fontsize=6)
plt.title("Languages Used in 2024")
plt.show()
```

- To rotate use startangle = 90 ( Not only 90 , you can use any value between 0 and 360)
- Change labels font sizes use textprops = {'fontsize': 10} in the legend function.
- counterClock = False ( To rotate the labels in the opposite direction)
