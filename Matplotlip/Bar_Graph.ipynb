### Just Example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [6,5,4,3,2]
c = ['r','g','b','y','r']

plt.bar(x,y, color=c)
plt.xlabel("Count_NO")
plt.ylabel("Lable_NO")
```

## 1. Matplotlib Bar Plot

```python
x = ['Ali','Sumi','Rup','Sumon']
y = [90,80,40,65]
z = [95,20,70,10]
width = 0.4

p = np.arange(len(x))
p1 = [j+width for j in p]

plt.bar(p,y, width, label = "2023", align="edge")
plt.bar(p1,z,width, color = 'r', label = "2024", align="edge")
plt.xlabel("Name")
plt.ylabel("No")
plt.legend(fontsize = 8)

plt.xticks(p+width,x)
plt.show()
```

### Same but Horizontal Bar graph

```python
lan = ['python','c','c++','java']
pop = [90,60,70,80]
pop2 = [56,78,23,28]

width = 0.2

p = np.arange(len(lan))

p2 = [j+width for j in p]

# Figure
plt.figure(figsize=(6,5))

# X lable
plt.xlabel("Language", fontsize = 10)


# Y lable
plt.ylabel("Popularity")

# Title
plt.title("Language Popularity", fontsize = 10)

# Create Bar
plt.barh(p,pop, width, align="edge", label = "Popularity")
plt.barh(p2,pop2, width, color = "y",align="edge", label = "Popularity_2")



# Show lan name
plt.yticks(p+width,lan)


# To Show legend
plt.legend(fontsize = 8)

# Show Bar
plt.show()
```

```python
name = ['Ali','Python',"Java","C++"]
n = [90,78,56,49]
nn = [89,98,60,20]

width = 0.2

name0 = np.arange(len(name))
name1 = [width+j for j in name0]



plt.barh(name0,n,width, align="edge", label = "Year 2022")
plt.barh(name1, nn, width, align="edge",label ="Year 2023")

plt.legend()
plt.title("ali_yt_98")
plt.xlabel("Name")
plt.ylabel("Popularity")


plt.yticks(name0+width, name)
plt.show()
```

```python
x = ['Ali','Sumi','Rup','Sumon']
y = [90,80,40,65]
z = [95,20,70,10]

width = 0.4

p = np.arange(len(x))
p1 = [width+j for j in p]


plt.bar(p,y,width,align="center", label = "Number_0")
plt.bar(p1, z, width, align="center",label = "Number_1")

plt.xticks(p+width, x)
plt.legend()
plt.show()
```
