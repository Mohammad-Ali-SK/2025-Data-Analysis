# 7. Merge, Join and Concatenate in Pandas

```python
data1 = {"Emp ID": ["E01","E02","E03","E04","E05"],
         "Name":["Ali","Sumi","Rup","Roki","Sahin"],
         "Age":[22,45,66,77,40]}

data2 ={"Emp ID": ["E01","E02","E03","E04","E05"],
         "Salary": [33000,56000,60000,34000,78000]}

df1 = pd.DataFrame(data1)
df2  = pd.DataFrame(data2)
df1


```

```python
df2
```

```python
pd.merge(df1, df2, on='Emp ID')
```
