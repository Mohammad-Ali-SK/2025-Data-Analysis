## Handling Missing Value in Gender

```python
new_df['Gender'] = new_df['Gender'].fillna('Male')
```

## Label Encoding

```python
le = preprocessing.LabelEncoder()
new_df['Gender_label'] = le.fit_transform(new_df.Gender.values)
```

```python
new_df.drop(columns=['Exited'], inplace=True)
```

```python
labels = ['0-20','21-40','41-60','Above 61']
bins = [0,20,40,60,100]
new_df['Age_labels'] = pd.cut(new_df.Age, bins, labels=labels, include_lowest=True)
```

## One HOT Encoding

```python
## Use of Pandas

one_hot = pd.get_dummies(new_df['Geography'])

```

```python
df_dummies = pd.get_dummies(new_df)
df_dummies.head()
```

```python
df_dummies = pd.get_dummies(new_df)

#  convert boolean to integer

df_dummies = df_dummies.astype(int)

df_dummies.head()
```

## Dummy Encoding

```python
df_dummies_de = pd.get_dummies(new_df, drop_first=True)
df_dummies_de = df_dummies_de.astype(int)
df_dummies_de.head()
```
