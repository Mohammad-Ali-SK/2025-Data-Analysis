## Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.set_style("whitegrid")
orders = ["Sat", "Sun", "Thur", "Fri"]
sns.countplot(x = "day", data = df, order=orders, palette="rocket")
plt.show()
```

```python
sns.violinplot(data=df, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", fill=False,
               palette="rocket")

plt.xlabel("Day Name")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")
sns.set_palette("mako")

# Create a violin plot with enhanced readability and visualization
plt.figure(figsize=(10, 6))  # Set figure size
ax = sns.violinplot(
    data=df,
    x="day",
    y="total_bill",
    hue="smoker",
    split=True,
    inner="quart",
    fill=False,
    linewidth=1.2,  # Adjust line width for better visibility
)

# Customize labels and title
ax.set_xlabel("Day of the Week", fontsize=12, fontweight="bold")
ax.set_ylabel("Total Bill Amount ($)", fontsize=12, fontweight="bold")
ax.set_title("Total Bill Distribution by Day and Smoking Status", fontsize=14, fontweight="bold")

# Improve legend positioning
plt.legend(title="Smoker", title_fontsize=12, fontsize=10, loc="upper right")

# Display the plot
plt.show()
```
