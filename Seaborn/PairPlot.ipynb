# Pair Plot in Seaborn

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.set_style("ticks")
sns.set_palette("rocket")
sns.pairplot(data = df, hue = "sex")

plt.show()
```

### To Create only particular type of data

```python
sns.pairplot(data = df, vars=['total_bill',"tip"], hue='sex', markers=["s","D"])
plt.show()
```

- IF you want to change type of plot then use kind = "reg" ( You use another value like = "scatter", "reg", "hist", "kde")
- To change Markers use markers = "o" (You use another value like = "o", "s", "D")
