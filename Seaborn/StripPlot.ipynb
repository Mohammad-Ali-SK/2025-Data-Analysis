## Strip Plot in Python

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
sns.set_style("dark")
sns.stripplot(x = "day", y="tip", data = df, palette="rocket", hue="sex")
plt.show()
```

- Use Size parametter to change size
- Use linewidth to change linewidth
