## Box Plot in Python

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
# Set Seaborn style
sns.set_palette("rocket")

# Create boxplot: total_bill vs. day
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=df)

# Show plot
plt.show()
```

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Create the boxplot with enhancements
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=df, palette="coolwarm")

# Add title and labels
plt.title("Total Bill Distribution by Day (With Smoker/Non-Smoker Comparison)", fontsize=14, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Total Bill (USD)", fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=11)

# Add legend with title
plt.legend(title="Smoker", title_fontsize="13", fontsize=11)

# Show plot
plt.show()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(8,6))

sns.boxplot(x = "day", y = "total_bill", data = df, palette="coolwarm", hue="sex", notch=True)

plt.xlabel("Day of week", fontsize = 12)
plt.ylabel("Total Bill", fontsize = 12)
plt.title("Total Bill Distribution by Day SEX", fontsize=14, fontweight='bold')
plt.legend(title="SEX", loc = 2)
plt.show()
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Create the boxplot for total_bill only
plt.figure(figsize=(6, 6))
sns.boxplot(y=df["total_bill"], color="royalblue", width=0.4, notch=False)

# Add title and labels
plt.title("Distribution of Total Bill Amounts", fontsize=14, fontweight='bold')
plt.ylabel("Total Bill (USD)", fontsize=12)

# Show plot

plt.show()
```
