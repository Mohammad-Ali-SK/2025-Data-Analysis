# CountPlot in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("penguins")
```

```python
data.head()
```

```python
sns.countplot(data['sex'])
plt.show()
```

```python
# Set style
sns.set_style("whitegrid")

# Create figure
plt.figure(figsize=(6, 5))

# Count plot with vertical bars
sns.countplot(x="sex", data=data, palette="pastel", edgecolor="black")

# Add labels and title
plt.xlabel("Sex", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Distribution of Sex", fontsize=14, fontweight="bold")

# Show plot
plt.show()
```

```python
# Set style
sns.set_style("whitegrid")

# Create figure
plt.figure(figsize=(6, 5))

# Create count plot
ax = sns.countplot(x="sex", data=data, palette="pastel", edgecolor="black")

# Add values on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2, p.get_height()),  
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Add labels and title
plt.xlabel("Sex", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Distribution of Sex", fontsize=14, fontweight="bold")

# Show plot
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(6,5))

ax = sns.countplot(x = 'sex', data = data, palette="pastel", edgecolor = "black",)

plt.xlabel("Sex",fontsize = 12)
plt.ylabel("Count",fontsize = 12)
plt.title("Sex Distribution", fontsize = 14, fontweight = "bold")

plt.show()
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,5))

sns.countplot(x = "sex", data = df, edgecolor = "black", palette="bwr", hue="smoker")
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize = 12)
plt.title("Gender Distribution", fontsize = 14, fontweight = "bold")
plt.show()
```
