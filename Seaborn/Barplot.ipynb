# BarPlot in Python Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head(5)
```

```python
df['island'].unique()
```

```python
sns.barplot(x = "island",y=df['bill_length_mm'], data = df, hue="sex")
plt.legend(loc = 2)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island",y=df['bill_length_mm'], data = df, hue="sex", order=order_1, hue_order=['Female','Male'], palette="hls",
            saturation=0.8, errcolor="black")
plt.legend(loc = 2)
plt.show()
```

- If you want to change axis to use -- orient = "h"/"v"
- Change color use palette = "viridis"/"plasma"/"inferno"/"magma"/"cividis
- Use saturation to change color saturation use -- saturation = 0.5/1.0
- Change errcolor to change error color use -- errcolor = "red"/"blue"/"green"/"yellow
- Use alpha to change transparency use -- alpha = 0.5/1.0
