## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("penguins").head(50)
```

```python
sns.scatterplot(x=df['bill_length_mm'], y = df['bill_depth_mm'], data=df)
plt.show()
```

#### After using Chat GPT

```python
# Set a stylish theme
sns.set_style("whitegrid")

# Create scatter plot with better customization
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", color="blue", edgecolor="black", alpha=0.7, palette="hls")

# Add labels and title
plt.xlabel("Bill Length (mm)", fontsize=12)
plt.ylabel("Bill Depth (mm)", fontsize=12)
plt.title("Relationship Between Bill Length and Bill Depth", fontsize=14, fontweight="bold")

# Show the plot
plt.show()
```

```python
# Set a stylish theme
sns.set_style("whitegrid")

# Create figure
plt.figure(figsize=(8, 6))

# Scatter plot
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", color="blue", edgecolor="black", alpha=0.7, hue="sex", style="sex", size="sex", palette="hls")

# Regression line
sns.regplot(data=df, x="bill_length_mm", y="bill_depth_mm", scatter=False, color="red", line_kws={"linewidth": 2})

# legend location
plt.legend(loc='upper right')

# Add labels and title
plt.xlabel("Bill Length (mm)", fontsize=12)
plt.ylabel("Bill Depth (mm)", fontsize=12)
plt.title("Relationship Between Bill Length and Bill Depth", fontsize=14, fontweight="bold")

# Show the plot
plt.show()
```
