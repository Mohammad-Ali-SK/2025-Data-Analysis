## Histogram Plot in Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head(2)
```

```python
df['bill_depth_mm'].unique()
```

```python
sns.displot(df['bill_depth_mm'],kde = True)
```

- To Organize data use bins
- If show line use kde  = True

```python
# Set the style for better visualization
sns.set_style("whitegrid")

# Create the distribution plot with enhancements
plt.figure(figsize=(8, 5))
sns.histplot(df['bill_depth_mm'], bins=[0, 10, 12, 14, 16, 18, 20, 22], kde=True, color="royalblue", edgecolor="black")

# Add labels and title
plt.xlabel("Bill Depth (mm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Bill Depth (mm)", fontsize=14)
plt.savefig("bill_depth_distribution.png", dpi=300, bbox_inches="tight")

# Show the plot
plt.show()
```

```python
# Set style for better visualization
sns.set_style("whitegrid")

# Create figure and axis
plt.figure(figsize=(8, 5))

# Plot histogram with KDE
sns.histplot(df['bill_depth_mm'], bins=[13, 14, 16, 18, 20, 22], kde=True, color="royalblue", edgecolor="black", alpha=0.7)

# Make KDE line more visible
sns.kdeplot(df['bill_depth_mm'], color="darkblue", linewidth=2.5)

# Adjust x-axis limits to remove empty space
plt.xlim(13, 22)

# Add labels and title
plt.xlabel("Bill Depth (mm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Bill Depth (mm)", fontsize=14)

# Save the improved plot
plt.savefig("bill_depth_distribution_improved.png", dpi=300, bbox_inches="tight")

# Show the plot
plt.show()
```
